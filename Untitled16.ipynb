```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import StandardScaler
```

```python
def compare_clustering_performance(X, true_labels, algorithm, algorithm_name):
    predicted_labels = algorithm.fit_predict(X)
    ari = adjusted_rand_score(true_labels, predicted_labels)
    return ari
```

```python
diabetes_df = pd.read_csv("/content/diabetes.csv")
```

```python
diabetes_true_labels = diabetes_df.iloc[:, -1].values
```

```python
diabetes_X = diabetes_df.iloc[:, :-1].values
```

```python
scaler = StandardScaler()
diabetes_X_scaled = scaler.fit_transform(diabetes_X)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
hierarchical = AgglomerativeClustering(n_clusters=2)
```

```python
ari_diabetes_kmeans = compare_clustering_performance(diabetes_X_scaled, diabetes_true_labels, kmeans, "K-Means")
ari_diabetes_hierarchical = compare_clustering_performance(diabetes_X_scaled, diabetes_true_labels, hierarchical, "Hierarchical")
```

```python
print("\nAdjusted Rand Index for Diabetes dataset:")
print(f"K-Means: {ari_diabetes_kmeans}")
print(f"Hierarchical: {ari_diabetes_hierarchical}")
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(diabetes_X_scaled[:, 0], diabetes_X_scaled[:, 1], c=kmeans.fit_predict(diabetes_X_scaled), cmap='viridis')
plt.title('K-Means Clustering (Diabetes)')

plt.subplot(1, 2, 2)
plt.scatter(diabetes_X_scaled[:, 0], diabetes_X_scaled[:, 1], c=hierarchical.fit_predict(diabetes_X_scaled), cmap='viridis')
plt.title('Hierarchical Clustering (Diabetes)')

plt.show()
```
